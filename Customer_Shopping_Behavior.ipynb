```python
df['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

## FEATURE ENGINEERING

```python
# AGE GROUPING CUSTOMER BY AGE TO THIS FOLLOWING CATEGORY
labels = ['Young Adult', 'Adult', 'Middle-Aged Adult', 'Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(5)
```

```python
# Creating columns Purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(5)
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```
